```python
import pandas as pd
```

```python
foodies=pd.read_csv('2020-XTern-DS.csv')
```

```python
foodies.head()
```

```python
#Location description

foodies.describe()
```

```python
#highest rating

best_rating=[]
for rating in foodies['Rating']:
    if rating == 'NEW':
        continue
    if rating == '-':
        continue
    if rating == 'Opening Soon':
        continue
    best_rating.append(float(rating))
highest_rating=0.0
for rating in best_rating:
    if rating>=highest_rating:
        highest_rating=rating
highest_rating
```

```python
#highest votes

b = foodies["Votes"] == foodies['Votes'].max()
foodies[b]
```

```python
#top 5 most southern restaurants

lat = sorted(foodies['Latitude'])
nums = lat[:5]
n = foodies[foodies['Latitude'].isin(nums)]
# n=sorted(n['latitude'])
n=n.sort_values(by='Latitude')
south = n['Restaurant'].tolist()
south
```
